```python
n=int(input("Enter num:"))
digit1=n%10

square_num=n**2
digit2=square_num%10

if digit1==digit2:
  print("Automorphic Number")

else:
  print("Not Automorphic Number")

```

Harshad Number:
A number divisible by sum of its digits

```python
N=int(input("Enter num:"))
temp=N
sum=0
while temp>0:
  digit=temp%10
  sum=sum+digit
  temp=temp//10
if N%sum==0:
    print("Harshad")
else:
    print("Not Harshad")
```


7. Count Numbers with Even Digits Only
Input: N = 50
Count numbers whose digits are all even.

```python
N = 50
i = 1

while i <= N:
    num = i
    total_digits = 0
    even_digits = 0

    while num > 0:
        digit = num % 10
        total_digits += 1

        if digit % 2 == 0:
            even_digits += 1

        num = num // 10

    if total_digits == even_digits:
        print(i)

    i += 1
```

4. Reverse Until Condition
Keep reversing a number until it becomes a palindrome.




```python
num = int(input("Enter number: "))

rev = int(str(num)[::-1])

while num != rev:
    num = num + rev
    print("New number:", num)
    rev = int(str(num)[::-1])

print("Palindrome:", num)
```

1. Strong Number Check
A number is called strong if sum of factorial of digits = number
👉 Example: 145 = 1! + 4! + 5!
Task: Input a number and check if it's a strong number.



```python
num = int(input("Enter number: "))
temp = num
sum_fact = 0

while temp > 0:
    digit = temp % 10

    # factorial of digit
    fact = 1
    i = 1
    while i <= digit:
        fact *= i
        i += 1

    sum_fact += fact
    temp = temp // 10

if sum_fact == num:
    print("Strong Number")
else:
    print("Not Strong Number")
```

6. Stop at First Prime
Print numbers from 1 to N
👉 Stop the loop when first prime number appears (break)


```python
N = int(input("Enter N: "))
i = 2   # start from 2 (first prime)

while i <= N:
    j = 2
    is_prime = True

    while j < i:
        if i % j == 0:
            is_prime = False
            break
        j += 1

    if is_prime:
        print(i)
        break

    i += 1
```
