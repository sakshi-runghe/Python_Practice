PYTHON PRACTICE QUESTIONS



```python
#1.Print numbers from 1 to 10 except 3 and 8
for i in range(1,10):
  if i==3 or i==8:
    continue
  print(i)
```

```python
#2.Find the first number divisible by 7 between 1 and 100
for i in range(1,101):
  if i%7==0:
    print(i)
    break

```

```python
#3.count vowels in a string
str_ = "aaeiooudmsgnanhejksaaioua"
vowels = "aeiou"

count = 0
for ch in str_:
    if ch in vowels:
        count += 1

print(count)
```

```python
#4.print all numbers from 1 to 30 except multiples of 5
for i in range(1,30):
  if i %5==0:
    continue
  print(i)
```

```python
#5.print numbers 1-50 but skips multiples of 3 ,skip numbers ending with 7 ,stop completely if number becomes 37
for i in range(1, 51):
    if i == 37:      # stop completely
        break
    if i % 3 == 0 or i % 10 == 7:  # skip conditions
        continue
    print(i)
```

```python
#6.remove duplicates elements from tuple
tuple_=(2,3,4,2,2,7,2)
x=set(tuple_)
print(tuple(x))
```

```python
#7.find the second largest element without using sort() in a list
list1=[3,2,3,23,67,45]
list1.sort()
list1
print(list1[-2])
```

```python
#8.split a list into two halves
lst = [1, 2, 3, 4, 5, 6]

mid = len(lst) // 2
first = lst[:mid]
second = lst[mid:]

print(first)
print(second)
```

```python
#9.find common elements between two lists without using set()
a = [1, 2, 3, 4, 5]
b = [3, 4, 6, 7]
common = []
for x in a:
    if x in b and x not in common:
        common.append(x)
print(common)
```

```python
#10.check if list is palindrome
lst = [1, 2, 3, 2, 1]
if lst == lst[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")

```

```python
#11.count frequency of each character
s = "SakShi"
count = {}

for ch in s:
    if ch in count:
        count[ch] += 1
    else:
        count[ch] = 1
print(count)

```

```python
#12.find duplicate characters
s = "programming"
for ch in s:
    if s.count(ch) > 1:
        print(ch)
```

```python
#13.remove duplicate characters
s = "programming"
result = ""

for ch in s:
    if ch not in result:
        result += ch
print(result)
```

```python
#14.replace vowels with *
s = "programming"
vowels = "aeiouAEIOU"
result = ""

for ch in s:
    if ch in vowels:
        result += "*"
    else:
        result += ch
print(result)
```

```python
#15.count words in a sentence
s = "Welcome to the world!"

words = s.split()
print(words)
print(len(words))

```
